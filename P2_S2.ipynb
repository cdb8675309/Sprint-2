```python
import pandas as pd
import numpy as np
red_wine = pd.read_csv('/Users/cd/Documents/Code Academy Berlin/Project 2/wine+quality/winequality-red.csv', sep = ';')
red_wine_copy = red_wine.copy()
white_wine = pd.read_csv('/Users/cd/Documents/Code Academy Berlin/Project 2/wine+quality/winequality-white.csv', sep = ';')
white_wine_copy = white_wine.copy()
```

```python
def add_underscores(red_wine):
    red_wine.columns = red_wine.columns.str.replace(' ', '_')
    return red_wine
```

```python
red_wine = add_underscores(red_wine)
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
```

```python
# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
def add_underscores(white_wine):
    white_wine.columns = white_wine.columns.str.replace(' ', '_')
    return white_wine
```

```python
white_wine = add_underscores(white_wine)
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
```

```python
# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'],
categories=['low', 'medium', 'high']) 
```

```python
#here i make a column titled red_or_white in the white_wine list, and give all the white wines property 'white'

white_wine['red_or_white'] = 'white'

white_wine
```

```python
#here i make a column titled red_or_white in the red_wine list, and give all the red wines property 'red'


red_wine['red_or_white'] = 'red'

red_wine
```

```python
combiwines = pd.concat([red_wine, white_wine], ignore_index=True)
combiwines
```
